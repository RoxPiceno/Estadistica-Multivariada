```python
#e.tree es libreria dentro de python para usar arboles de raíz hacia

import xml.etree.ElementTree as ET

# Create a dummy XML file for demonstration
xml_data = '''
<movielist>
    <movie>
        <title>The Shawshank Redemption</title>
        <director>Frank Darabont</director>
        <year>1994</year>
    </movie>
    <movie>
        <title>The Godfather</title>
        <director>Francis Ford Coppola</director>
        <year>1972</year>
    </movie>
    <movie>
        <title>Pulp Fiction</title>
        <director>Quentin Tarantino</director>
        <year>1994</year>
    </movie>
</movielist>
'''
#'w' es para sólo lectura

with open('movielist.xml', 'w') as f:
    f.write(xml_data)

print("Dummy 'movielist.xml' created successfully.")

# Parse the XML file abrir y recorder un archivo
tree = ET.parse('movielist.xml')
root = tree.getroot()

# Iterate and print movie data
print("\nMovie DataBase")
for movie in root.findall("movie"): #desde la raíz encuentra cada hijo
    title = movie.find("title").text
    director = movie.find("director").text
    year = movie.find("year").text
    print(f"Title: {title}, Director: {director}, Year: {year}")
```
