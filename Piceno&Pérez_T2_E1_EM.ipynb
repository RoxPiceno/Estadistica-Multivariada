```python
# Regresi√≥n Polin√≥mica en Python
#Ejercicio 1 sobre Predicci√≥n del PIB

# Paso_1: Importar librer√≠as, siempre al principio del c√≥digo y s√≥lo las que se usar√°n.

import pandas as pd
import statsmodels.api as sm  #para regresi√≥n m√∫ltiple
import matplotlib.pyplot as plt


#Se ponen en comillas simples para identificar las palbras como etiquetas porque es un string, o sea datos.
# Y el programa necesita saber que son strings para construir el DataFrame correctamente

#Paso_2: crear los datos manualmente para este ejemplo

datos = {
   'PIB': [600, 1200, 1800, 2500, 3100, 3700, 4200, 4800, 5300, 5900],
   'inversion': [50, 100, 150, 200, 250, 300, 350, 400, 450, 500],
   'exportaciones': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
}
#Se usan las llaves {} para crear diccionarios. Un diccionario es una estructura que guarda pares
# y una colecci√≥n de varios pares se encierra entre llaves; [ ] ‚Üí significa lista, ( ) ‚Üí significa tupla o funciones, { } ‚Üí significa diccionario

#convertir a DataFrame

df = pd.DataFrame(datos)

#Denominar las variables
# Y= PIB (variable dependiente), X= Exportaciones e Inversi√≥n (variable independientes)
# se usa df para nombrar la variable, aunque se puede usar cualquier palabra, se acostumbra
# df por DataFrame, que es la estructura principal de pandas para manejar datos en forma de tabla
# Y Un DataFrame funciona como una tabla parecida a Excel columnas (filas, encabezados, index, c√°lculos, filtrado, selecci√≥n de columnas)

###Paso_3: Definir las variables, dependiente e independiente.

Y = df['PIB']
X = df[['inversion', 'exportaciones']]

###Paso_4: Ajustar el modelo con statsmodels; a√±adir la constante

# En una regresi√≥n lineal, la ecuaci√≥n del modelo completo es: PIB=Œ≤0‚Äã+Œ≤1‚Äã(inversion)+Œ≤2‚Äã(exportaciones)+Œµ ; Œ≤0= intercepto o constante; Œ≤1= coeficiente de inversi√≥n; Œ≤2= coeficiente de exportaciones
# La constante es lo que predice el modelo cuando todas las variables valen 0.
# Peroooo statsmodels NO agrega la constante autom√°ticamente si se pasasn los datos as√≠ como est√°n, statsmodels solo estima los coeficientes, pero no el intercepto.
# sklearn (m√°s orientado a machine learning, r√°pido y simple pero no da todo lo necesario) SI la agrega autom√°ticamente, pero statsmodels es m√°s estad√≠stico, te da p-values, R¬≤, resumen completo
# Para un ejercicio econ√≥mico (PIB, inversi√≥n, exportaciones), lo m√°s adecuado es statsmodels porque en econom√≠a interesa:
# Significancia estad√≠stica, intervalos de confianza, p-values, interpretaci√≥n tipo ‚Äúpor cada 1 unidad de inversi√≥n, el PIB sube
# por eso es necesario agregar la constante a X antes de ajustar el modelo

X = sm.add_constant(X)  #Es el intercepto en la regresi√≥n

# calcular los coeficientes de la regresi√≥n (calcula los valores √≥ptimos de ùõΩ0,ùõΩ1,ùõΩ2)
# primero se debe crear el objeto del modelo: esto NO corre la regresi√≥n solo prepara el modelo. Despu√©s se hace el ajuste.

#crear modelo OLS
modelo = sm.OLS(Y, X).fit()
print (modelo.summary())

###Paso_5: interpretar coeficientes

###Paso_6: predecir PIB

nuevos_datos = pd.DataFrame({
    'const': [1],
    'inversion': [550],
    'exportaciones': [1100]
})

prediccion = modelo.predict(nuevos_datos)
print("Predicci√≥n de PIB:", prediccion.values[0])

```
